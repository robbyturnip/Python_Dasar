```python
import pandas as pd

anime_df = pd.read_csv('dataanime-norm.csv')
anime_df.head(10)
```

```python
anime_df[anime_df['Starting season'] == 'Fall']
```

```python
anime_df[anime_df['Genres'].str.contains('Action')==True]
```

```python
from matplotlib import pyplot as plt

anime_trend = anime_df[['Start airing','Members']]
anime_trend = anime_trend.sort_values(by='Start airing',ascending=True)

plt.plot(anime_trend['Start airing'],anime_trend['Members'])
plt.locator_params(axis='x',nbins=10)
plt.xticks(rotation=90)
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Penonton')
plt.title('Trend Jumlah Penonton')
plt.show()
```

```python
from matplotlib import pyplot as plt

anime_trend = anime_df[['End airing','Favorites']]
anime_trend = anime_trend.sort_values(by='End airing',ascending=True)

plt.plot(anime_trend['End airing'],anime_trend['Favorites'])
plt.locator_params(axis='x',nbins=10)
plt.xticks(rotation=90)
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Penonton')
plt.title('Trend Jumlah Penonton')
plt.show()
```
