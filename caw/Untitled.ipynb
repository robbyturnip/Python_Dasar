```python
import numpy as np

fall_df=baca_anime[baca_anime["Starting season"]=="Fall"]
spring_df=baca_anime[baca_anime["Starting season"]=="Spring"]
diagram.plot(np.log(fall_df["Members"]),np.log(fall_df["Favorites"]),"ob")
diagram.plot(np.log(spring_df["Members"]),np.log(spring_df["Favorites"]),"or")
diagram.show()
```

```python
from scipy.stats.stats import pearsonr
print(pearsonr(fall_df["Members"],fall_df["Favorites"]))
print(pearsonr(spring_df["Members"],spring_df["Favorites"]))
```

```python
fall_trend = fall_df.sort_values(by='Start airing',ascending=True)
diagram.plot(fall_trend['Start airing'],fall_trend['Members'])
diagram.show()
```

```python
new_fall_trend = fall_trend['Members'].rolling(window=20).mean()
diagram.plot(fall_trend['Start airing'],new_fall_trend)
diagram.show()
```

```python
new_fall_trend = fall_trend['Members'].rolling(window=200).mean()
diagram.plot(fall_trend['Start airing'],new_fall_trend)
diagram.show()
```

```python
#SIMPLE MOVING AVERAGE
new_fall_trend = fall_trend['Members'].rolling(window=20).mean()
diagram.plot(fall_trend['Start airing'],new_fall_trend)

#EXPONENTIAL MOVING AVERAGE
exp_fall_trend = fall_trend['Members'].ewm(span=20,adjust=False).mean()
diagram.plot(fall_trend['Start airing'],exp_fall_trend,'r')
diagram.show()
```
